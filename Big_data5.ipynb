```python
from sklearn.datasets import load_iris
```

```python
iris_df = pd.DataFrame(data.data, columns= data.feature_names)
```

```python
iris_df['encoded_target'] = data.target
```

```python
# retirei do dataset original a coluna cm e espaços entre intems da tabela e coloquei underline para as colunas ficarem mais code friendly
```

```python
iris_df.columns = [c.replace("(cm)","").rstrip().replace(" ","_") for c in iris_df.columns]
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar','baz', 'foo'],
                    'value': [1,2,3,4]})
```

```python
df2 = pd.DataFrame ({'rkey': ['foo', 'bar','baz', 'foo'],
                     'value': [5,6,7,8]})
```

```python
print(df1.merge(df2, left_on='lkey', right_on='rkey', suffixes=("_left","_right")))
```
